```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Salah_all_PL_shots.csv',encoding='windows-1252')
print(data.head())
```

```python
plt.figure(figsize=(10,8))

palette = {result: 'green' if result == 'Goal' else 'red' for result in data['result'].unique()}

sns.scatterplot(x=data['X'], y=data['Y'], hue=data['result'], palette=palette)

plt.title('Map of Mo Salah shots')
plt.show()
```
